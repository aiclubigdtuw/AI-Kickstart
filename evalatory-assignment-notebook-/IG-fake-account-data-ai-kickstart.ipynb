## 3. An expressive or an non-expressive user?

An user is said to be expreessive if he/she has a description length above then average, but first we need to know about the average value of a description length from our data.

So we use .describe method

```python
data_train.describe()
```

## --More clearly the mean is --

```python
data_train['description length'].mean(axis=0)
```

> We get to know that the average length of an instagram bio should be greater than 22 words, max is 150 words and min is 0 words. Taking this as our basis we can divide our data into : non-expressive if desc_len is in between 0- 22 and expressive if desc_len is in between 22-150.

```python
data_exp_non_exp = data_train
data_exp_non_exp["Expressive?"] = np.where(data_exp_non_exp['description length'] >22,'Expressive','Non-Expressive')
```

```python
data_exp_non_exp
```

```python
data_exp_non_exp[data_exp_non_exp["Expressive?"] == "Expressive"].shape
```

### Thus 182 people arefound to be more expressive then the average.  

```python
fig, axes = plt.subplots(2, 2, figsize=(18,20), sharey=False)
fig.suptitle('The coorelation of expressiveness of an user depending upon- Desc_len, fake, private, no. of posts \n',size=15)


sns.barplot(ax=axes[0][0] ,x="Expressive?",y="description length",hue="Expressive?",data=data_exp_non_exp,palette="Set2")
sns.barplot(ax=axes[1][1] ,x="Expressive?",y="fake",hue="Expressive?",data=data_exp_non_exp,palette="Set2")
sns.barplot(ax=axes[1][0] ,x="Expressive?",y="private",hue="Expressive?",data=data_exp_non_exp,palette="Set2")
sns.barplot(ax=axes[0][1] ,x="Expressive?",y="#posts",hue="Expressive?",data=data_exp_non_exp,palette="Set2")
```

### From this representation we see that if a person is expressive he/she also has more number of posts, is lower chances of having account private and also lower chances of being a fake user.
