## EVALUATORY PROJECT

```python
#importing the standard libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon = pd.read_csv("pokemon.csv")
```

```python
pokemon.shape
```

```python
pokemon.head()
```

```python
y = pokemon['target'].values
```

```python
X = pokemon.drop(columns=['target'], inplace=False).values
```

```python
y.shape
```

```python
X.shape
```

```python
plt.imshow(X[300].reshape(100,100,3))
```

```python
sns.countplot(x='target',data=pokemon)
```

```python
#splitting the data into train and test datasets

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
print( X_train.shape, y_train.shape )
```

```python
print( X_test.shape, y_test.shape)
```

## KNN ALGORITHM

```python
# importing the KNeighborsClassifier

from sklearn.neighbors import KNeighborsClassifier
```

```python
# creating an object of KNeighborsClassifier with k=3

model = KNeighborsClassifier(n_neighbors=3)
```

```python
# model training on X_train and y_train

model.fit(X_train, y_train)
```

```python
#Predictions on X_test part

model.predict(X_test)
```

### Evaluation

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
print(f"KNN ALGORITHM SCORE:{model.score(X_test, y_test)}")
```

### Predicting for any random image

```python
for i in [10, 20, 75, 10, 65, 21, 30, 70, 25, 46]:
    random_img = X_test[i]
    plt.imshow(random_img.reshape(100,100,3))
    plt.show()
    print("Actual Label : ",y_test[i])
    p = model.predict( [random_img] )
    print("Predicted Label : ", p)
```

## LOGISTIC REGRESSION

```python
# importing the logistic regression model

from sklearn.linear_model import LogisticRegression
```

```python
# creating an object of Logistic Regression Model

model = LogisticRegression()
```

```python
# model training on X_train and y_train

model.fit(X_train, y_train)
```

```python
#Predictions on X_test part
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

### Evaluation

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
print(f"LOGISTIC REGRESSION SCORE:{model.score(X_test, y_test)}")
```

### Predicting for any random image

```python
for i in [69, 24, 75, 19, 65, 21, 56, 70, 20, 46]:
    random_img = X_test[i]
    plt.imshow(random_img.reshape(100,100,3))
    plt.show()
    print("Actual Label : ",y_test[i])
    p = model.predict( [random_img] )
    print("Predicted Label : ", p)
```
