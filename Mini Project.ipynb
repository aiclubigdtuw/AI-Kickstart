### QUES 5:  Fake accounts have been witnessed to heavily skew the follower and following numbers. What is the ratio of number of followers (#followers) to number of following (#follows) for the user with the maximum number of following?

```python
foll=max(test['#followers'])
max_fol=test[test['#followers']==foll]
print("profile of the user with maximum number of following")
max_fol
```

```python
# calculating the ratio 
a_1=max(max_fol['#followers'])
print("number of followers of the user: "+str(a_1))
a_2=max(max_fol['#follows'])
print("number of follows of the user: "+str(a_2))
```

```python
r1=a_1/a_2
```

```python
r1
```

```python
# function to calculate the ratio
def ratio(d):
    b=d.as_integer_ratio()
    return b
b=ratio(r1)
print("Ratio is: ")
print(b[0], ":", b[1])  

#the approx ratio is about 729:1
```

#### Data Visualisation

```python
print("#followers VS #follows  for the person who has maximum following")
x=np.array([a_1,a_2])
mylabels=['#followers','#follows']
mycolors=['blue','hotpink']
plt.pie(x,colors=mycolors)
plt.legend(mylabels)
plt.show()
```
