```python

#Importing Python Modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as 
```

```python

# Reading pokemon.csv

pokémon = pd.read_csv('pokemon.csv')
```

```python

pokémon.shape
```

```python

# Overviewing the file

pokémon.head(2)
```

```python

# Assigning the Pokémon Names to y
y = pokémon['target'].values

# Assigning the Pixels_Info to X
X = pokémon.drop(columns = ['target'], inplace = False).values
```

```python

# y and X information

print("X.shape :", X.shape)
print("y.shape :", y.shape)
```

```python

# Step to Display Poké_Identity

plt.imshow(X[200].reshape(100, 100, 3))
```

```python

# Importing function from sklearn.model_selection module to split our data

from sklearn.model_selection import train_test_split
```

```python

# Splitting our Data into Train and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python

# Info of Train and Test Sets


print("X_train.shape :", X_train.shape)
print("y_train.shape :", y_train.shape)
print()
print("X_test.shape :", X_test.shape)
print("y_test.shape :", y_test.shape)
```

```python

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 7)
```

```python

# Training our model with Train set

model.fit(X_train, y_train)
```

```python

# Predicting the Test set based on Training

y_pred = model.predict(X_test)

print("Predictions based on Training:-\n\n")
print(y_pred)
print("\n\n")
```

```python

#Accuracy of Prediction Model

acc = model.score(X_test, y_test)
print(acc)
```

# Accuracy: 0.7540983606557377  == 75.40% (approx)
